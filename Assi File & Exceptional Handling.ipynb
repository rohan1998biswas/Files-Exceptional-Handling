```python
import threading

lis=[]
lislock=threading.Lock()

def add_num():
    for i in range(10):
        time.sleep(0.5)
        with lislock:
            lis.append(i)
            print(f"Added: {i}, List: {lis}")

def remove_num():
    for i in range(10):
        time.sleep(0.5)
        with lislock:
            if lis:
                removed_val=lis.pop(0)
                print(f"Removed: {removed_val},List: {lis}")

add_thread=threading.Thread(target=add_num)
remove_thread=threading.Thread(target=remove_num)
add_thread.start()
remove_thread.start()
add_thread.join()
remove_thread.join()
print("Final List:",lis)
```

```python
# 7. Create a program that uses a thread pool to calculate the factorial of numbers from 1 to 10 concurrently. Use concurrent.futures.ThreadPoolExecutor to manage the threads.
```

```python
import concurrent.futures

def cal_factorial(n):
    res=1
    for i in range(1,n+1):
        res*=i
    return res

num=range(1,11)
with concurrent.futures.ThreadPoolExecutor() as e:
    res=e.map(cal_factorial,num)
for num,factorial in zip(num,res):
    print(f"Factorial of {num} is {factorial}")
```

```python
# 8. Create a Python program that uses multiprocessing.Pool to compute the square of numbers from 1 to 10 in parallel. Measure the time taken to perform this computation using a pool of different sizes (e.g., 2, 4, 8 processes)
```

```python
import multiprocessing
import time

def square(n):
    return n**2

def comp_squares(pool_size):
    num=range(1,11)
    with multiprocessing.Pool(pool_size) as p:
        res=p.map(square,num)
    return res

pool_sizes=[2, 4, 8]
for i in pool_sizes:
    start=time.perf_counter()
    res=comp_squares(i)
    end=time.perf_counter()
    print(f"Pool Size: {i}, Results: {res}, Time Taken: {round(end-start, 2)} seconds")
```
